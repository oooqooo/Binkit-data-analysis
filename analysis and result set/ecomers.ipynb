```python
df["order_date"]=pd.to_datetime(df["order_date"])
```

```python
df["year"]=df["order_date"].dt.year
df["day"]=df["order_date"].dt.day
df["qurter"]=df["order_date"].dt.quarter
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["sales"]=df["sales"].astype(int)
```
